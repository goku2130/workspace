## Clustering of Products based on Time Series - KMeans

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans_kwargs = {
    "init" : "random",
    "n_init" : 10,
    "max_iter" : 300,
    "random_state" : 123
}
sse = []
for k in range(1, 11):
    kmeans= KMeans(n_clusters = k, **kmeans_kwargs)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.axvline(3, label = 'Best cluster count', c = 'r')
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.legend()
plt.title('Elbow Method for Number of Clusters')
plt.show()
```

```python
kmeans.cluster_centers_[0]
```

```python
from sklearn.metrics import silhouette_score

silhouette_coefficients = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(data)
    score = silhouette_score(data, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.axvline(3, label = 'Best cluster count', c = 'r')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.legend()
plt.title('Silhouette Coefficient Method for Number of Clusters')
plt.show()
```

```python
kmeans= KMeans(n_clusters = 3, **kmeans_kwargs)
kmeans.fit(data)
data['cluster'] = kmeans.labels_
data['cluster'] = data['cluster'].astype(str)
```

```python
import statsmodels.api as sm
ts = pd.DataFrame()

data = group_1a.loc['P788'].to_frame()
data = data.reset_index().iloc[0:52]
data['P788'] =data['P788'].astype(int)
data.plot()
```

```python
sm.graphics.tsa.plot_acf(data['P788'], lags=20)
plt.show()
```

```python
sm.graphics.tsa.plot_pacf(data['P788'], lags=20)
plt.show()
```

The ACF and PACF plots look like that of white noise. Also, there is no trend or seasonality in the time series. 

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
def adf_test(timeseries):
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
adf_test(data['P788'])
```

We can safely reject the null hypothesis of ADF test, the time series is stationary. We can conclude by running KPSS test to see if there is any difference.

```python
def kpss_test(timeseries):
    print ('Results of KPSS Test:')
    kpsstest = kpss(timeseries, regression='c', nlags="auto")
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    print (kpss_output)
```

```python
kpss_test(data['P788'])
```

We failed to reject the null hypothesis for the KPSS test. After the tests, we can say that the time series is stationary.

This group contains products that do not have any sort of pattern in them. It would be better to know more about the description of these products to understand the demand cycle. But, given the information that we have, we cannot make any forecast.

## Forecasting Model for Group 1a (Simple Exponential Smoothing)

Since the time series for the product does not have a trend and is stationary, we can use Simple Exponential Smoothing model. It gives a flat forecast and gives more importance to recent events.

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit1 = SimpleExpSmoothing(data['P788'], initialization_method="heuristic").fit(smoothing_level=0.2,optimized=False)
fcast1 = fit1.forecast(3).rename(r'$\alpha=0.2$')
fit2 = SimpleExpSmoothing(data['P788'], initialization_method="heuristic").fit(smoothing_level=0.6,optimized=False)
fcast2 = fit2.forecast(3).rename(r'$\alpha=0.6$')
fit3 = SimpleExpSmoothing(data['P788'], initialization_method="estimated").fit()
fcast3 = fit3.forecast(3).rename(r'$\alpha=%s$'%fit3.model.params['smoothing_level'])
```

```python
plt.figure(figsize=(12, 8))
line0, =plt.plot(data['P788'], color='black', label = 'true')
plt.plot(fit1.fittedvalues, color='blue')
line1, = plt.plot(fcast1,marker='o', color='blue')
plt.plot(fit2.fittedvalues, color='red')
line2, = plt.plot(fcast2,marker='o', color='red')
plt.plot(fit3.fittedvalues, color='green')
line3, = plt.plot(fcast3, marker='o', color='green')
plt.legend([line0, line1, line2, line3], ['true', fcast1.name, fcast2.name, fcast3.name])
plt.show()
```

```python
fit3.summary()
```

The Simple Exponential Smoothing model is optimized to chose the best value of alpha. The optimized value of alpha is 0. It is expected as the stationary time series is fluctuating around the mean of 3.4.  Effectively, the same model can be applied to all other products belonging to this category. For all the products belonging to Group 1a, we can procure 52-wk average transaction count per week.

## Forecasting Model for Group 2 - ARIMA(2, 1, 1) vs ARIMA(2, 1, 0)

```python
data = group_2.loc['P46'].to_frame()
data = data.reset_index().iloc[0:52]
data['P46'] = data['P46'].astype(int)
data.dtypes
```

```python
from statsmodels.tsa.arima.model import ARIMA
arma_mod = ARIMA(data['P46'], order=(2, 1, 1))
arma_res = arma_mod.fit()
print(arma_res.summary())
```

```python
from statsmodels.graphics.tsaplots import plot_predict
fig, ax = plt.subplots(figsize=(10,8))
plt.plot(data[['P46']], color ='black', label = 'true')
arma_res.fittedvalues.plot(label='fitted value')
fig = plot_predict(arma_res, start = 52, end = 55, ax=ax)
legend = ax.legend(loc='upper left')
plt.title('Actual vs Forecast for P46 using ARIMA(2, 1, 1) model including forecasts for next 3 weeks')
plt.xlabel('Weeks')
```

```python
import numpy as np
np. set_printoptions(suppress=True)
print('P-values for various lags in ljung box test for serial autocorrelation on residuals:',\
      np.round(arma_res.test_serial_correlation('ljungbox', lags=None), 3)[0][1])
```

Since the residuals are serially uncorrelated (tested using Ljung Box test and failed to reject the null), can conclude that the model is a good fit. Let us try a less complex model, ARIMA(2, 1, 0)

```python
from statsmodels.tsa.arima.model import ARIMA
arma_mod = ARIMA(data['P46'], order=(2, 1, 0))
arma_res = arma_mod.fit()
print(arma_res.summary())
```

```python
print('P-values for various lags in ljung box test for serial autocorrelation on residuals:',\
      np.round(arma_res.test_serial_correlation('ljungbox', lags=None), 3)[0][1])
```

```python
fig, ax = plt.subplots(figsize=(10,8))
plt.plot(data[['P46']], color ='black', label = 'True')
arma_res.fittedvalues.plot(label='fitted value')
fig = plot_predict(arma_res, start = 52, end = 55, ax=ax)
legend = ax.legend(loc='upper left')
plt.title('Actual vs Forecast for P46 using ARIMA(2, 1, 0) model including forecasts for next 3 weeks')
plt.xlabel('Weeks')
```

Since the BIC score is lesser for ARIMA(2, 1, 0) as compared to ARMA(2, 1, 1), we would go with a ARIMA(2, 1, 0) model for forecasting the products of this group 2.

```python
from statsmodels.tsa.exponential_smoothing.ets import ETSModel

model = ETSModel(data['P334'].astype(float), trend='add', damped_trend=True)
ets_res = model.fit()
plt.figure(figsize=(12, 8))
plt.plot(data[['P334']], color ='black', label = 'true')
ets_res.fittedvalues.plot(label='fitted value')
pred = ets_res.get_prediction(start=52, end=55)
pred = pred.summary_frame(alpha=0.05)
pred["mean"].plot(label='mean prediction')
pred["pi_lower"].plot(linestyle='--', color='tab:blue', label='95% interval')
pred["pi_upper"].plot(linestyle='--', color='tab:blue', label='_')
plt.ylabel("Tx count")
plt.xlabel('Week')
plt.title('Time Series - Product P334')
plt.legend()
```
